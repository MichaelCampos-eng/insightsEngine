```python
user_profiles = pd.DataFrame()
original_posts = pd.DataFrame()
social_medias = pd.DataFrame()
posts_infos = pd.DataFrame()
```

```python
for data in user_infos:
    user_profile = data["user_profile"]
    original_post = data['original_post']
    social_media = data['social_media']
    posts_info = data['posts_info']
    author_id = user_profile["author_uid"]
    
    original_post["author_id"] = author_id
    social_media["author_id"] = author_id
    
    for post_info in posts_info:
        post_info["author_id"] = author_id
        posts_infos = pd.concat([posts_infos, pd.DataFrame([post_info])], ignore_index=True)
    
    social_medias = pd.concat([social_medias, pd.DataFrame([social_media])], ignore_index=True)
    original_posts = pd.concat([original_posts, pd.DataFrame([original_post])], ignore_index=True)
    user_profiles = pd.concat([user_profiles, pd.DataFrame([user_profile])], ignore_index=True)
```
